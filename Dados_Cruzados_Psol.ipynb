```python
import pandas as pd
import numpy as np
```

```python
bzs = pd.read_csv("BAIRROS_ZONA_SECAO2.csv", sep=";", encoding="utf-8")
```

```python
votos = pd.read_csv("urnas_votos_psol_2018.csv", sep=";", usecols=['NR_ZONA','NR_SECAO','NM_VOTAVEL','QT_VOTOS','QT_APTOS', 'QT_COMPARECIMENTO','QT_ABSTENCOES','SG_PARTIDO'], encoding="ISO-8859-1")
```

```python
bzs
```

```python
votos
```

```python
votos_geral = pd.merge(votos, bzs,
                  how='inner',
                 on=['NR_ZONA', 'NR_SECAO'])
# result.to_csv('testecruza.csv' , sep=';', encoding ="ISO-8859-1")
```

```python
votos_alex = (votos_geral['NM_VOTAVEL'] == 'PROF ALEX FRAGA')
alex = votos_geral[votos_alex]
```

```python
alex
```

```python
votos_robaina= (votos_geral['NM_VOTAVEL'] == 'ROBERTO ROBAINA')
robaina = votos_geral[votos_robaina]
```

```python
x = robaina[['QT_VOTOS']].groupby(robaina['BAIRRO'])
```

```python
y = x.sum()
```

```python
y.to_csv('votos_robaina_2018.csv', sep = ';', encoding = 'ISO-8859-1')
```
