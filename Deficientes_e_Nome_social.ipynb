```python
import pandas as pd
```

```python
bzs = pd.read_csv('eleitorado/base_pesquisa/BAIRROS_ZONA_SECAO.csv', sep=';', encoding='utf-8')
```

```python
elei = pd.read_csv('eleitorado/base_pesquisa/perfil_eleitor_secao_2020_RS.csv', sep = ';',encoding="ISO-8859-1", usecols =['NM_MUNICIPIO' , 'NR_ZONA', 'NR_SECAO','DS_GENERO', 'DS_ESTADO_CIVIL', 'DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE', 'QT_ELEITORES_PERFIL' , 'QT_ELEITORES_INC_NM_SOCIAL', 'QT_ELEITORES_DEFICIENCIA'])
```

## Seleção de Eleitores de Porto Alegre

```python
selecao_poa = (elei['NM_MUNICIPIO'] == 'PORTO ALEGRE') & (elei['QT_ELEITORES_INC_NM_SOCIAL'] == 1) & (elei['DS_GENERO'] == 'MASCULINO')
```

```python
elei_poa = elei[selecao_poa]
```

```python
elei_bairro = pd.merge(elei_poa, bzs,
                  how='inner',
                 on=['NR_ZONA', 'NR_SECAO'])
```

```python
faixa_etaria = ['BAIRRO', 'QT_ELEITORES_INC_NM_SOCIAL'] 
ele = elei_bairro.filter(items=faixa_etaria)
```

```python
ele['QT_ELEITORES_INC_NM_SOCIAL'].value_counts()
ele.groupby('BAIRRO').count().to_csv('trans_masc.csv', sep = ';', encoding="ISO-8859-1")
```

## Seleção de Eleitores Deficientes

```python
selecao_deficientes =  (elei_bairro['QT_ELEITORES_DEFICIENCIA'] > 0)
```

```python
elei_deficientes = elei_bairro[selecao_deficientes]
elei_deficientes['DS_GENERO'].value_counts().to_csv('deficiente_genero.csv', sep= ';', encoding="ISO-8859-1")
```

```python
faixa_etaria = ['BAIRRO', 'QT_ELEITORES_DEFICIENCIA'] 
ele = elei_bairro.filter(items=faixa_etaria)
```

```python
ele.groupby('BAIRRO').count().to_csv('deficiente_feminino.csv', sep = ';', encoding="ISO-8859-1")
```

```python
grau_escolaridade = ['DS_GRAU_ESCOLARIDADE', 'QT_ELEITORES_DEFICIENCIA'] 
ele = elei_bairro.filter(items=grau_escolaridade)
```

```python
ele.groupby('DS_GRAU_ESCOLARIDADE').count().to_csv('deficiente_escolaridade.csv', sep = ';', encoding="ISO-8859-1")
```
