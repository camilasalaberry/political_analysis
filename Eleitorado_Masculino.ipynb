```python
import pandas as pd
```

```python
bzs = pd.read_csv('eleitorado/base_pesquisa/BAIRROS_ZONA_SECAO.csv', sep=';', encoding='utf-8')
```

```python
elei = pd.read_csv('eleitorado/base_pesquisa/perfil_eleitor_secao_2020_RS.csv', sep = ';',encoding="ISO-8859-1", usecols =['NM_MUNICIPIO' , 'NR_ZONA', 'NR_SECAO','DS_GENERO', 'DS_ESTADO_CIVIL', 'DS_FAIXA_ETARIA', 'DS_GRAU_ESCOLARIDADE', 'QT_ELEITORES_PERFIL' , 'QT_ELEITORES_INC_NM_SOCIAL', 'QT_ELEITORES_DEFICIENCIA'])
```

```python
selecao = (elei['NM_MUNICIPIO'] == 'PORTO ALEGRE') & (elei['DS_GENERO'] == 'MASCULINO')
```

```python
elei_poa = elei[selecao]
```

```python
elei_poa.sort_values(ascending=False, by=['QT_ELEITORES_PERFIL'])
```

```python
elei_bairro = pd.merge(elei_poa, bzs,
                  how='inner',
                 on=['NR_ZONA', 'NR_SECAO'])
```

```python
elei_bairro.sort_values(ascending=False, by=['QT_ELEITORES_PERFIL'])
```

```python
elei_bairro.sort_values(ascending=False, by=['QT_ELEITORES_PERFIL'])

selecao = ['DS_GRAU_ESCOLARIDADE', 'DS_FAIXA_ETARIA', 'DS_ESTADO_CIVIL', 'BAIRRO'] 
ele = elei_bairro.filter(items=selecao)
```

```python
ele['DS_GRAU_ESCOLARIDADE'].value_counts(normalize=True)*100
```

```python
ele['DS_FAIXA_ETARIA'].value_counts().to_csv('test_ms_idade.csv', sep = ';', encoding="ISO-8859-1")
```

```python
ele['DS_ESTADO_CIVIL'].value_counts(normalize=True)*100
```
